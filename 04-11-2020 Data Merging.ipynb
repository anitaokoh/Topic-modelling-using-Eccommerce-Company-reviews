```python
new_score = score.max(axis=1)
```

```python
sentiment_data['Score'] = list(new_score.values)
```

```python
sentiment_data.head()
```

```python
def priority_status(x):
    if x < 0.5:
        y = 'Low-priority'
    elif x <= 0.7:
        y = 'Moderate-priority'
    else:
        y='High-priority'
    return y
```

```python
sentiment_data['priority'] = sentiment_data['Score'].apply(priority_status)
```

```python
sentiment_data['Sentiment'] = np.where(sentiment_data['Sentiment'].isna(),'NEGATIVE',sentiment_data['Sentiment'])
```

```python
sentiment_data.drop(['SentimentScore','ErrorCode','ErrorMessage'],axis=1,inplace=True)
```

```python
sentiment_data.sort_values(by=['Line'],inplace=True)
sentiment_data = sentiment_data.iloc[1:,:]
sentiment_data['Line'] = range(0,653)
sentiment_data.reset_index(inplace=True,drop=True)
```
